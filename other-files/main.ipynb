```python
#Funcion GetWarehouses
def GetWarehouses(apikey: str) -> pd.DataFrame:
    QUERY_PARAMETERS = {'APIKEY':apikey}
    URL = 'https://app.stelorder.com/app/warehouses'
    response = requests.get(URL,params=QUERY_PARAMETERS)
    dt_frame = pd.DataFrame(response.json())
    return dt_frame
```
